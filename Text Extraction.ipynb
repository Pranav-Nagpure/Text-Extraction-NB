## Text Extraction from Images

```python
import requests
import pytesseract

from PIL import Image
```

#### Getting image

```python
img = Image.open(requests.get('https://media.geeksforgeeks.org/wp-content/cdn-uploads/20220401160946/HTML-Basic-Format-768x534.png', stream=True).raw)
img.resize((500, 300))
```

#### Converting to Grayscale

```python
img = img.convert('L')
img.resize((500, 300))
```

#### Extracting Text

```python
# Extracting Text
custom_config = r'-l eng --oem 3 --psm 6'
text = pytesseract.image_to_string(img, config=custom_config)
print(text)
```

```python
# Removing Special Characters
text = ''.join(c for c in text if (c.isalnum() or c == ' ' or c == '\n'))
print(text)
```
